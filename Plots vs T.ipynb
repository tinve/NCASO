```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
#import os.path
#from random import randint
#from random import seed
```

```python
fnames = ['2.5D, 6x6x2 spins.csv',
          '2.5D, 8x8x2 spins.csv',
          '2.5D, 10x10x2 spins.csv',
          '2.5D, 12x12x2 spins.csv',
          '2.5D, 20x20x2 spins.csv']

hue_order = [fname[:-4] for fname in fnames]
hue_order
```

```python
data = pd.concat([pd.DataFrame.from_csv(fname) for fname in fnames])

data
```

```python
sns.factorplot('T', 'E', hue = 'type', hue_order = hue_order, kind = 'point',
                legend_out = False, data = data)
sns.plt.show()
```
